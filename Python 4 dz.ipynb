## Нижняя граница аномальных значений:
### для quest - 10.000.000
### для mob_reward	- 100.000.000

```python
def cheater_detecter(dataframe):
    return dataframe[((dataframe['source_class'] == 'quest') & (dataframe['value'] >= 10000000)) 
                  | ((dataframe['source_class'] == 'mob_reward') & (dataframe['value'] >= 100000000))]
    
```

```python
cheater_detecter(data)
```
