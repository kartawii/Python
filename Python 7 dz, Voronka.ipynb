### Задание 2

```python
install = pd.read_csv('install.csv')
tutorial_complete = pd.read_csv('tutorial_complete.csv')
match_start = pd.read_csv('match_start.csv')
purchase = pd.read_csv('purchase.csv')
```

### Очищаю от полностью одинаковых строк

```python
install = install.drop_duplicates()
tutorial_complete = tutorial_complete.drop_duplicates()
match_start = match_start.drop_duplicates()
purchase = purchase.drop_duplicates()
```

### Очищаю одинаковые Id, чтоб не портили статистику

```python
tutorial_complete = tutorial_complete['user_id'].drop_duplicates().reset_index(drop=True)
match_start = match_start['user_id'].drop_duplicates().reset_index(drop=True)
purchase = purchase['user_id'].drop_duplicates().reset_index(drop=True)
```

### Вывожу значения воронки

```python
print('В выборке было',len(install),'установок')
print('Доля прошедших туториал:', len(tutorial_complete) / len(install))
print('Доля сыгравших первый матч:', len(match_start) / len(install))
print('Доля совершивших первую покупку:', len(purchase) / len(install))
```
