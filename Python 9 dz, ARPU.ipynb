## Задание 1

```python
arrpu_dataset = pd.read_csv('arrpu_dataset.csv',index_col = 0)
```

```python
def count_arpu(df):
    arpu = df['ARPPU']*df['PPU']
    return arpu

arrpu_dataset['ARPU'] = arrpu_dataset.apply(count_arpu,axis=1)
arrpu_dataset
```

## Задание 2

```python
logins = pd.read_csv('les9_logins.csv',index_col = 0)
logins
```

```python
def level_group(lvl):
    if lvl < 6 :
        return 'low'
    if 5 < lvl < 11 :
        return 'mid'
    if 10 < lvl < 21:
        return 'high'
    return 'core_players'

logins['level_group'] = logins['current_lvl'].apply(level_group)
logins
```

```python
logins['event_timestamp'] = pd.to_datetime(logins['event_timestamp'], format='%Y-%m-%d')

logins['day_start'] = logins['event_timestamp'].dt.to_period('d').dt.start_time
logins
```

```python
logins = logins.groupby(['level_group','day_start']).agg({'client_id':'count'})
```

```python
logins = logins.groupby(['level_group']).agg({'client_id':'mean'})
logins.columns = ['AVG']
logins
```
