## Задание 1

```python
import pandas as pd

import numpy as np
```

## Посмотрим на данные

```python
data = pd.read_csv('logins_by_server.csv')
data
```

## Избавимся от ютуберов и тестеров

```python
data = data[(data['is_tester'] == 0) & (data['is_youtuber'] == 0)]
data
```

## Вычислим кол-во уникальных айди

```python
data_n = data.groupby(['event_date','game_server'], as_index=False).agg({'client_id':'nunique'})
data_n
```

## Посчитаем среднее для каждого сервера

```python
data_s = data_n.groupby(['game_server'], as_index=False).agg({'client_id':'mean'})
data_s
```

## Выведем плохие сервера

```python
bad_server = data_s[(data_s['client_id']) < 100]
bad_server
```
